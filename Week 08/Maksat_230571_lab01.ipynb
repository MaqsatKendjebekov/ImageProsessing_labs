```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# If running in Google Colab, uncomment below to upload an image
# from google.colab import files
# uploaded = files.upload()

# Load example image from internet
import urllib.request

url = 'https://static.tildacdn.com/tild6530-6465-4134-a165-373062643564/Centralasianuz_news_.jpg'
urllib.request.urlretrieve(url, 'image.png')

image = cv2.imread('image.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
gx = np.zeros_like(image)
gy = np.zeros_like(image)

gx[:, :-1] = np.diff(image, axis=1)
gy[:-1, :] = np.diff(image, axis=0)

gradient_magnitude = np.sqrt(gx**2 + gy**2)

plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Simple Gradient Magnitude')
plt.axis('off')
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')
```

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F)

plt.imshow(np.abs(laplacian), cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')
```

```python
canny = cv2.Canny(image, 100, 200)

plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detection with 100, 200')
plt.axis('off')
```

```python
canny = cv2.Canny(image, 200, 400)

plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detection with 200, 400')
plt.axis('off')
```

```python
canny = cv2.Canny(image, 400, 800)

plt.imshow(canny, cmap='gray')
plt.title('Canny Edge Detection with 400, 800')
plt.axis('off')
```

```python
plt.figure(figsize=(16,12))
Gaussian = cv2.GaussianBlur(image, (15, 15), 0)  
laplacian = cv2.Laplacian(Gaussian, cv2.CV_64F)
sobel_x = cv2.Sobel(Gaussian, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(Gaussian, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)


plt.subplot(1,3,1)
plt.imshow(Gaussian, cmap='gray')
plt.title('Gaussian blured')
plt.axis('off')


plt.subplot(1,3,2)
plt.imshow(np.abs(laplacian), cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')



plt.subplot(1,3,3)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')
```
