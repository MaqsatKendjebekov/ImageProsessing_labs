```python
import cv2
import matplotlib.pyplot as plt

color_image = cv2.imread('1.png')

(b, g, r) = cv2.split(color_image)

equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)

equalized_color_image = cv2.merge([equalized_b,equalized_g, equalized_r])

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(color_image,
cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_color_image,
cv2.COLOR_BGR2RGB))
plt.title('Equalized Color Image (ChannelsSeparately)')
plt.axis('off')
plt.show()
```

```python
import cv2 
import numpy as np
from skimage import exposure 
import matplotlib.pyplot as plt


ims = cv2.imread('1.png', cv2.IMREAD_GRAYSCALE)
imt = cv2.imread('bright.png', cv2.IMREAD_GRAYSCALE)


mimg = exposure.match_histograms(ims, imt)

plt.figure(figsize=(18,10))

plt.subplot(3,3,1)
plt.imshow(ims, cmap='gray')
plt.title('Source Image')
plt.axis('off')

plt.subplot(3,3,2)
plt.hist(ims.ravel (), 256, [0,256])
plt.title('Histogram (Source) ')

plt.subplot(3,3,4)
plt.imshow(imt, cmap='gray')
plt.title('Target Image')
plt.axis('off')

plt.subplot(3,3,5)
plt.hist(imt.ravel(),256,[0,256])
plt.title('Histogram (Target)')

plt.subplot(3,3,7)
plt.imshow(mimg,cmap='gray')
plt.title('Matched Image')
plt.axis('off')

plt.subplot(3,3,8)
plt.hist(mimg.ravel(),256,[0,256])
plt.title('Histogram (Matched) ')

plt.show()
```
