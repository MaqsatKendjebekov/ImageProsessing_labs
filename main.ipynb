```python
import cv2


c = cv2.imread('CAU.png')
c_img = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
cv2.imshow('BGR2RGB',c_img)

cv2.waitKey(0)
```

```python
import cv2

g = cv2.imread('CAU.png', cv2.IMREAD_GRAYSCALE)

_, b_img = cv2.threshold(g, 128,255,cv2.THRESH_BINARY)
cv2.imwrite("img1.png",b_img)

cv2.waitKey(0)
```

```python
import cv2
g = cv2.imread('CAU.png', cv2.IMREAD_GRAYSCALE)
_, b_img = cv2.threshold(g, 100,150,cv2.THRESH_BINARY)
cv2.imwrite("img2.png", b_img)

cv2.waitKey(0)
```
