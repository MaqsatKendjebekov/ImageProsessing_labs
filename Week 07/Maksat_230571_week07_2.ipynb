```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show(img, title="", cmap="gray", size=(5,5)):
    plt.figure(figsize=size)
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

img_square = np.zeros((200, 200), dtype=np.uint8) 
img_circle = np.zeros((200, 200), dtype=np.uint8)

 
plt.figure(figsize=(16,16))

cv2.circle(img_circle, (100, 100), 50, 255, -1)

cv2.rectangle(img_square, (50, 50), (150, 150), 255, -1)



sobel_x = cv2.Sobel(img_circle, cv2.CV_64F, 1, 0, ksize=1)
sobel_y = cv2.Sobel(img_circle, cv2.CV_64F, 0, 1, ksize=1)

s_sobel_x = cv2.Sobel(img_square, cv2.CV_64F, 1, 0, ksize=1)
s_sobel_y = cv2.Sobel(img_square, cv2.CV_64F, 0, 1, ksize=1)



plt.subplot(1,3,1)
plt.title("Img Circle")
plt.imshow(img_circle)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Img Sobel X")
plt.imshow(np.abs(sobel_x))
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Img Sobel Y")
plt.imshow(np.abs(sobel_y))
plt.axis("off")


plt.subplot(3,3,1)
plt.title("Img Square")
plt.imshow(img_square)
plt.axis("off")

plt.subplot(3,3,2)
plt.title("Img Sobel X")
plt.imshow(np.abs(s_sobel_x))
plt.axis("off")

plt.subplot(3,3,3)
plt.title("Img Sobel Y")
plt.imshow(np.abs(s_sobel_y))
plt.axis("off")


```

```python

for k in [3, 5, 7]:
    sobel_k = cv2.Sobel(img_circle, cv2.CV_64F, 1, 0, ksize=k)
    show(np.abs(sobel_k), f"Sobel X on circle - ksize={k}")
```

```python
img_step = np.zeros((200, 200), dtype=np.uint8)

img_step[:, :100] = 50
img_step[:, 100:] = 220
show(img_step, "Step image")
```
