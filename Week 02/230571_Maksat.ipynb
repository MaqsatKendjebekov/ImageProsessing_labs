```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('CAU.png')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#image becomes brighter because in RGB((255+100),(255+100),(255+100))
added_image = cv2.add(image_RGB, 100)

plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(added_image)
plt.title('Added Image (100)')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('CAU.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#less than 1 we will get darker more than 1 brighter
mul_image = cv2.multiply(image_rgb,0.5)

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(mul_image)
plt.title('Multiplied Image 2')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('CAU.png')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

mulimg = cv2.divide(imgRGB, 2)

plt.subplot(1, 2, 1)
plt.imshow(imgRGB)
plt.title("original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(mulimg)
plt.title('Divided img')
plt.axis('off')

plt.show()
```
