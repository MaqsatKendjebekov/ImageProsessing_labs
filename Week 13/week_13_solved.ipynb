## Solved tasks
- **Task 1:** Compare morphology with kernel sizes 3x3 and 5x5.
- **Task 2:** Compare THRESH_BINARY and THRESH_BINARY_INV.
- **Task 3:** Test segmentation on multiple sample images.

### Solution step 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

images = [data.coins(), data.page(), data.camera()]
k3 = np.ones((3, 3), np.uint8)
k5 = np.ones((5, 5), np.uint8)
```

### Solution step 2

```python
results = []
for img in images:
    _, th = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
    _, th_inv = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
    open3 = cv2.morphologyEx(th, cv2.MORPH_OPEN, k3)
    open5 = cv2.morphologyEx(th, cv2.MORPH_OPEN, k5)
    results.append((img, th, th_inv, open3, open5))
```

### Solution step 3

```python
img, th, th_inv, open3, open5 = results[0]
fig, ax = plt.subplots(1, 5, figsize=(16, 4))
ax[0].imshow(img, cmap='gray'); ax[0].set_title('Original')
ax[1].imshow(th, cmap='gray'); ax[1].set_title('Binary')
ax[2].imshow(th_inv, cmap='gray'); ax[2].set_title('Binary inv')
ax[3].imshow(open3, cmap='gray'); ax[3].set_title('Open 3x3')
ax[4].imshow(open5, cmap='gray'); ax[4].set_title('Open 5x5')
for a in ax: a.axis('off')
plt.tight_layout()
```
