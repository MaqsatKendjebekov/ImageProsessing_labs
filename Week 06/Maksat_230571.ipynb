```python
#TASK 1
import cv2 
import numpy as np
from skimage import exposure 
import matplotlib.pyplot as plt


ims = cv2.imread('dataset/barbara.bmp', cv2.IMREAD_GRAYSCALE)
imt = cv2.imread('dataset/fruits.png', cv2.IMREAD_GRAYSCALE)

kernel = np.array([[-1,-2,-1], [0,1,0],[1,2,1]])
imc = cv2.filter2D(ims,-1,kernel)

mimg = exposure.match_histograms(imc, imt)

plt.figure(figsize=(18,10))

plt.subplot(3,3,1)
plt.imshow(imc, cmap='gray')
plt.title('Source Image')
plt.axis('off')

plt.subplot(3,3,2)
plt.hist(imc.ravel (), 256, [0,256])
plt.title('Histogram (Source) ')

plt.subplot(3,3,4)
plt.imshow(imt, cmap='gray')
plt.title('Target Image')
plt.axis('off')

plt.subplot(3,3,5)
plt.hist(imt.ravel(),256,[0,256])
plt.title('Histogram (Target)')

plt.subplot(3,3,7)
plt.imshow(mimg,cmap='gray')
plt.title('Matched Image')
plt.axis('off')

plt.subplot(3,3,8)
plt.hist(mimg.ravel(),256,[0,256])
plt.title('Histogram (Matched) ')

plt.show()
#Because target image has more contrast and we are equalizing it 
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def gauss_filt(n, sigma=1.0):
    if n <= 0 or n % 2 == 0:
        raise ValueError("Window size must be apositive odd integer.")
    ax = np.linspace(-(n - 1) / 2., (n - 1) / 2.,n)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-0.5 * (np.square(xx)+np.square(yy))/np.square(sigma))
    kernel = kernel / np.sum(kernel)
    return kernel

image = cv2.imread('dataset/peppers_color.tif',cv2.IMREAD_GRAYSCALE)
fil_imagex3 = cv2.GaussianBlur(image, (3, 3),0)
fil_imagex5 = cv2.GaussianBlur(image, (5, 5),0)

plt.figure(figsize=(16,12))


plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis("off")


plt.subplot(1,3,2)
plt.imshow(fil_imagex3, cmap='gray')
plt.title('Filtred 3x3')
plt.axis("off")



plt.subplot(1,3,3)
plt.imshow(fil_imagex5, cmap='gray')
plt.title('Filtred 5x5')
plt.axis("off")
plt.show()
```
