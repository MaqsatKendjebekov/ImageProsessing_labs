## Solved tasks
- **Task 1:** Convert between RGB and CMY using custom functions.
- **Task 2:** Segment colors in HSV space.
- **Task 3:** Enhance color channels and compare outputs.

### Solution step 1

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import data

def rgb_to_cmy(image):
    img = image.astype(np.float32) / 255.0
    return 1.0 - img

def cmy_to_rgb(image):
    return np.clip((1.0 - image) * 255, 0, 255).astype(np.uint8)

img = data.astronaut()
```

### Solution step 2

```python
cmy = rgb_to_cmy(img)
rgb_back = cmy_to_rgb(cmy)
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
mask_red = cv2.inRange(hsv, (0, 40, 40), (15, 255, 255))
enhanced = cv2.convertScaleAbs(img, alpha=1.1, beta=15)
```

### Solution step 3

```python
fig, ax = plt.subplots(1, 4, figsize=(16, 4))
ax[0].imshow(img); ax[0].set_title('Original RGB')
ax[1].imshow(cmy); ax[1].set_title('CMY view')
ax[2].imshow(rgb_back); ax[2].set_title('Back to RGB')
ax[3].imshow(mask_red, cmap='gray'); ax[3].set_title('HSV color slice')
for a in ax: a.axis('off')
plt.tight_layout()
```
