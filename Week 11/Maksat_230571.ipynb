```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.ndimage import binary_erosion, binary_dilation, binary_opening, binary_closing

img = np.zeros((12, 12), dtype=np.uint8)

# Main object
img[3:9, 3:9] = 1

# Small hole inside the object
img[5, 5] = 0

# Small foreground noise
img[1, 10] = 1

kernel = np.ones((3, 3), dtype=np.uint8)

print(kernel)
print("\n")
print(img)
print("\n")

plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='gray', vmin=0, vmax=1)
plt.title("Original Binary Image")
plt.axis("off")
plt.show()



```

```python
eroded = cv2.erode(img, kernel, iterations=1)

plt.figure(figsize=(5, 5))
plt.imshow(eroded, cmap='gray', vmin=0, vmax=1)
plt.title("Eroded Image")
plt.axis("off")
plt.show()

print("\n")
print(eroded)
```

```python
dilated = cv2.dilate(img, kernel, iterations=1)

plt.figure(figsize=(5, 5))
plt.imshow(dilated, cmap='gray', vmin=0, vmax=1)
plt.title("Dilated Image")
plt.axis("off")
plt.show()

print("\n")
print(dilated)
```

```python
opened = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=(5, 5))
plt.imshow(opened, cmap='gray', vmin=0, vmax=1)
plt.title("Opened Image")
plt.axis("off")
plt.show()

print("\n")
print(opened)
```

```python
closed = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(5, 5))
plt.imshow(closed, cmap='gray', vmin=0, vmax=1)
plt.title("Closed Image")
plt.axis("off")
plt.show()

print(closed)
print("\n")
```

```python
images = [img, eroded, dilated, opened, closed]
titles = ["Original", "Erosion", "Dilation", "Opening", "Closing"]

plt.figure(figsize=(15, 3))
for i in range(len(images)):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i], cmap='gray', vmin=0, vmax=1)
    plt.title(titles[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
img2 = np.zeros((15, 15), dtype=np.uint8)

# Large rectangle
img2[3:12, 3:12] = 1

# Add small holes
img2[5, 5] = 0
img2[8, 8] = 0

# Add a gap in the boundary
img2[3:5, 7] = 0

# Add noise
img2[1, 1] = 1
img2[13, 13] = 1

opened2 = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)
closed2 = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img2, cmap='gray', vmin=0, vmax=1)
plt.title("Original Example 2")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(opened2, cmap='gray', vmin=0, vmax=1)
plt.title("Opening")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(closed2, cmap='gray', vmin=0, vmax=1)
plt.title("Closing")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
task1 = np.zeros((20, 20), dtype=np.uint8)

cx, cy = 10, 10
r = 4

Y, X = np.ogrid[:20, :20]
circle = (X - cx)**2 + (Y - cy)**2 <= r**2
task1[circle] = 1


# 3. Isolated noise pixels
task1[1, 15] = 1
task1[17, 3] = 1

plt.figure(figsize=(5, 5))
plt.imshow(task1, cmap='gray', vmin=0, vmax=1)
plt.title("Task 1: Binary Image")
plt.axis("off")
plt.show()
```

```python


# 3x3 kernel
kernel = np.ones((3, 3), dtype=np.uint8)

# Apply operations
eroded = binary_erosion(task1, structure=kernel).astype(np.uint8)
dilated = binary_dilation(task1, structure=kernel).astype(np.uint8)
opened = binary_opening(task1, structure=kernel).astype(np.uint8)
closed = binary_closing(task1, structure=kernel).astype(np.uint8)

# Show results
fig, axes = plt.subplots(1, 5, figsize=(15, 5))

images = [task1, eroded, dilated, opened, closed]
titles = ["Original", "Erosion", "Dilation", "Opening", "Closing"]

for ax, img, title in zip(axes, images, titles):
    ax.imshow(img, cmap='gray', vmin=0, vmax=1)
    ax.set_title(title)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
#With the 5 × 5 kernel, all morphological operations produce stronger changes than with the 3 × 3 kernel because the structuring element covers a larger neighborhood.
#The operation that changed the objects the most was erosion, since it significantly shrank both the square and the circle-like object,
#and small details could disappear completely. The operation that removed noise best was opening, because it first eroded isolated foreground pixels and then restored the main objects.
#The operation that filled holes best was closing, since it expanded the objects and then shrank them back, which helped close small gaps and fill internal holes.
```

```python
task4 = np.zeros((20, 20), dtype=np.uint8)

# Two nearby objects with a one-pixel gap
task4[5:10, 3:7] = 1
task4[5:10, 8:12] = 1

kernel = np.ones((3, 3), dtype=np.uint8)

dilated_task4 = binary_dilation(task4, structure=kernel).astype(np.uint8)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

axes[0].imshow(task4, cmap='gray', vmin=0, vmax=1)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(dilated_task4, cmap='gray', vmin=0, vmax=1)
axes[1].set_title("After Dilation")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
task5 = np.zeros((20, 20), dtype=np.uint8)

# Object with a small hole inside
task5[4:15, 4:15] = 1
task5[9, 9] = 0

kernel = np.ones((3, 3), dtype=np.uint8)

closed_task5 = binary_closing(task5, structure=kernel).astype(np.uint8)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

axes[0].imshow(task5, cmap='gray', vmin=0, vmax=1)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(closed_task5, cmap='gray', vmin=0, vmax=1)
axes[1].set_title("After Closing")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python

# Load images
finger = cv2.imread('/content/finger.png', cv2.IMREAD_GRAYSCALE)
text = cv2.imread('/content/text.png', cv2.IMREAD_GRAYSCALE)

# Binary threshold
_, finger_bin = cv2.threshold(finger, 127, 255, cv2.THRESH_BINARY)
_, text_bin = cv2.threshold(text, 127, 255, cv2.THRESH_BINARY)

# Kernel
kernel = np.ones((3, 3), np.uint8)
kernel2 = np.ones((4, 4), np.uint8)

# Morphological operations
finger_open = cv2.morphologyEx(finger_bin, cv2.MORPH_OPEN, kernel)
double_open = cv2.morphologyEx(finger_open,cv2.MORPH_OPEN, kernel2)

text_close = cv2.morphologyEx(text_bin, cv2.MORPH_CLOSE, kernel)
double_close = cv2.morphologyEx(text_close, cv2.MORPH_CLOSE, kernel2)
# Display
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1)
plt.imshow(finger_bin, cmap='gray')
plt.title("Finger Original")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(finger_open, cmap='gray')
plt.title("Finger Opening")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(double_open, cmap='gray')
plt.title("X2 Opening")
plt.axis("off")


plt.subplot(2, 3, 4)
plt.imshow(text_bin, cmap='gray')
plt.title("Text Original")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(text_close, cmap='gray')
plt.title("Text Closing")
plt.axis("off")

plt.subplot(2, 3, 6)
plt.imshow(double_close, cmap='gray')
plt.title("X2 Closing")
plt.axis("off")



plt.tight_layout()
plt.show()
```
